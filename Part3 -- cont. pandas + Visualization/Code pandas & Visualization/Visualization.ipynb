```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.linspace(1,10,100)
```

```python
plt.plot(x) # line plot
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x),'*')
plt.plot(x,np.cos(x), 'o')
plt.show()
```

```python
a = np.array([1,2,3,4])
b = np.array([5,6,1,9])
plt.plot(a,b,linestyle = 'dotted',color = 'red')
```

```python
plt.plot(x)
#title
plt.title('line plot of x values')
# x label
plt.xlabel('x label title')
# y label
plt.ylabel('y label title')

plt.show()
```

```python
plt.plot(x,x, label = 'X values')
plt.plot(x,x**2, label = 'Quadratic X')
plt.plot(x,x**3, label = 'Qubic X')
plt.legend() # to show labels
plt.title('X values figure')
plt.show()
```

```python
plt.plot(x,x, label = 'X values',color = 'green')
plt.plot(x,x**2, label = 'Quadratic X',color = 'red')
plt.plot(x,x**3, label = 'Qubic X',color = 'purple')
plt.legend() # to show labels
plt.title('X values figure')
plt.show()
```

## sub plot

```python
#create first subplot
plt.subplot(2,2,1) #subplot(n rows, n cols, n plot)
plt.plot(x)

#create sec subplot
plt.subplot(2,2,2) #subplot(n rows, n cols, n plot)
plt.plot(np.sin(x))

#create third subplot
plt.subplot(2,2,3) #subplot(n rows, n cols, n plot)
plt.plot(np.cos(x))

#create fourth subplot
plt.subplot(2,2,4) #subplot(n rows, n cols, n plot)
plt.plot(x, x **2)
```
