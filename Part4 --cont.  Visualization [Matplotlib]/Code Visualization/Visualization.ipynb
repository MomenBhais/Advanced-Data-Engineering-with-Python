```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.linspace(1,10,100)
```

```python
plt.plot(x)
#title
plt.title('line plot of x values')
# x label
plt.xlabel('x label title')
# y label
plt.ylabel('y label title')

plt.show()
```

```python
plt.plot(x,x, label = 'X values')
plt.plot(x,x**2, label = 'Quadratic X')
plt.plot(x,x**3, label = 'Qubic X')
plt.legend() # to show labels
plt.title('X values figure')
plt.show()
```

```python
plt.plot(x,x, label = 'X values',color = 'green')
plt.plot(x,x**2, label = 'Quadratic X',color = 'red')
plt.plot(x,x**3, label = 'Qubic X',color = 'purple')
plt.legend() # to show labels
plt.title('X values figure')
plt.show()
```
